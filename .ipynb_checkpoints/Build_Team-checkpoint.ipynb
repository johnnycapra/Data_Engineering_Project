```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
import tkinter as tk
from tkinter import ttk, messagebox
```

```python
#create list of players and stats

# Read the DataFrames
rushing_df = pd.read_csv('output_data/rushing_cleaned.csv')
injury_df = pd.read_csv('output_data/injury_cleaned.csv')

# Perform left join on the "Player" column
player_data = pd.merge(rushing_df, injury_df, on='player', how='left')

# Display the merged DataFrame
player_data
```

```python
# Get Fantasy Football Team Names

# Initializing DF for team picks
team_df = pd.DataFrame(columns=[
    'Team Name',
    'QB',
    'RB',
    'WR',
    'TE',
])

team_df
```

```python
#establishing QB, RB, WR, TE lists

qb_names = rushing_df[rushing_df['pos'] == 'QB']['player']
rb_names = rushing_df[rushing_df['pos'] == 'RB']['player']
wr_names = rushing_df[rushing_df['pos'] == 'WR']['player']
te_names = rushing_df[rushing_df['pos'] == 'TE']['player']
```

```python
class TeamNamesGUI:
    def __init__(self):
        self.team_names = []

        self.root = tk.Tk()
        self.root.title("Team Names")

        # Text box and button to add team names
        self.team_name_entry = tk.Entry(self.root)
        self.team_name_entry.pack()

        add_team_button = tk.Button(self.root, text="Add Team", command=self.add_team)
        add_team_button.pack()

        # Listbox to display team names
        self.team_listbox = tk.Listbox(self.root)
        self.team_listbox.pack()

        # Button to finish entering team names
        finish_button = tk.Button(self.root, text="Finish", command=self.finish_entering_teams)
        finish_button.pack()

    def add_team(self):
        team_name = self.team_name_entry.get()
        if team_name:
            self.team_names.append(team_name)
            self.team_name_entry.delete(0, tk.END)
            self.update_team_listbox()

    def update_team_listbox(self):
        self.team_listbox.delete(0, tk.END)
        for team in self.team_names:
            self.team_listbox.insert(tk.END, team)

    def finish_entering_teams(self):
        self.root.destroy()  # Close the current GUI
        self.open_blank_gui()

    def open_blank_gui(self):
        blank_root = tk.Tk()
        blank_root.title("Blank GUI")
        # Add components or functionality as needed in the blank GUI

        blank_root.mainloop()



class BlankGUI:
    def __init__(self, team_names):
        self.team_names = team_names

        self.root = tk.Tk()
        self.root.title("Blank GUI")

        # Dropdown list for positions
        tk.Label(self.root, text="Select Position:").pack()
        self.position_var = tk.StringVar()
        positions = ['QB', 'RB', 'WR', 'TE']
        position_dropdown = ttk.Combobox(self.root, textvariable=self.position_var, values=positions, state='readonly')
        position_dropdown.pack()

        # Listbox to display players for the selected position
        self.players_listbox = tk.Listbox(self.root)
        self.players_listbox.pack()

        # Button to finish and close the GUI
        finish_button = tk.Button(self.root, text="Finish", command=self.finish)
        finish_button.pack()

        # Placeholder for player data
        self.player_data = {
            'QB': qb_names,  
            'RB': rb_names,
            'WR': wr_names,
            'TE': te_names
            # Add other positions with player lists
        }

        # Bind event to update players list when position changes
        self.position_var.trace_add('write', self.update_players_list)

    def update_players_list(self, *args):
        selected_position = self.position_var.get()
        if selected_position in self.player_data:
            players_for_position = self.player_data[selected_position]
            self.players_listbox.delete(0, tk.END)
            for player in players_for_position:
                self.players_listbox.insert(tk.END, player)

    def finish(self):
        # Add logic for finishing the GUI (e.g., saving selected players)
        self.root.destroy()

    def run(self):
        self.root.mainloop()

# Example usage
team_names = ['Team1', 'Team2', 'Team3']  # Replace with actual team names
team_names_gui = TeamNamesGUI()
team_names_gui.run()

blank_gui = BlankGUI(team_names)
blank_gui.run()
```
